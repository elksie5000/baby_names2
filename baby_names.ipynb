```python

import pandas as pd
import numpy as np
import altair as alt
from ipywidgets import interact, widgets
from IPython.display import display, Image
pd.set_option('display.max_columns', 1000)
alt.data_transformers.disable_max_rows()
path = "https://www.ons.gov.uk/file?uri=/peoplepopulationandcommunity/birthsdeathsandmarriages/livebirths/datasets/babynamesinenglandandwalesfrom1996/1996tocurrent/babynames1996to2021.xlsx"

df_boys = pd.read_excel("babynames1996to2021.xlsx", sheet_name=3)
df_girls = pd.read_excel("babynames1996to2021.xlsx", sheet_name=4)
```

```python
def prep_gender_frames(gender, df):
    row_with_name_index = df[df.iloc[:, 0] == 'Name'].index[0]
    #Pull out the columns
    columns = df_boys.iloc[6, :53]
    #Strip the top rows
    df = df.iloc[(row_with_name_index+1):,:]#The 1 is used to #remove the column data row
    df = df.replace("[x]", 0)
    df.columns=columns
    #Create gender column
    df['gender'] = gender
    return df
df_list = [] 
boys_and_girls = {"boy": df_boys, "girl": df_girls}
for key, val in boys_and_girls.items():   
    data = prep_gender_frames(key, val)
    df_list.append(data)
df = pd.concat(df_list).reset_index(drop=True)


```

```python
def adjust_rank(row):
    if row['Measure'] == 'Rank':
        if row['Number'] == 0:
            if row['Gender'] == 'boy':
                return 4790
            elif row['Gender'] == 'girl':
                return 5582
    return row['Number']

df['Value'] = df.apply(adjust_rank, axis=1)
```

```python
# Function to update name options based on the selected first letter
def update_name_options(change):
    new_options = sorted(df[df['first_letter'] == change['new']]['Name'].unique().tolist())
    names_widget.options = new_options

# Function to plot data based on parameters
def plot_data(first_letter=None, names=None, gender=None, measure='Rank'):
    # Filter data based on parameters
    filtered_data = df.copy()
    if first_letter:
        filtered_data = filtered_data[filtered_data['first_letter'] == first_letter]
    if names:
        filtered_data = filtered_data[filtered_data['Name'].isin(names)]
    if gender:
        filtered_data = filtered_data[filtered_data['Gender'].isin(gender)]
    if measure:
        filtered_data = filtered_data[filtered_data['Measure'] == measure]

    # Create the plot
    plot = alt.Chart(filtered_data).mark_line().encode(
        x=alt.X('Year:O'),
        y=alt.Y('mean(Value):Q', sort='descending'),
        color=alt.Color('Gender:N', scale=alt.Scale(domain=['boy', 'girl'], range=['blue', 'pink'])),
        tooltip=['Name', 'Gender', 'Value', 'Measure', 'Year']
    ).interactive()

    return plot

# Interactive widgets
first_letter_widget = widgets.Dropdown(options=[None] + sorted(df['first_letter'].unique().tolist()), description='First Letter:', disabled=False)
first_letter_widget.observe(update_name_options, names='value')

names_widget = widgets.SelectMultiple(options=sorted(df['Name'].unique().tolist()), description='Names:', disabled=False)
gender_widget = widgets.SelectMultiple(options=['boy', 'girl'], description='Gender:', disabled=False)
measure_widget = widgets.Dropdown(options=['Rank', 'Count'], description='Measure:', disabled=False)

interact(plot_data, first_letter=first_letter_widget, names=names_widget, gender=gender_widget, measure=measure_widget);
```
